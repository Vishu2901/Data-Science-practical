```python
import json
import time

fd = open("Records.json",'r')
js = fd.read()
fd.close()
record = json.loads(js)
user_in = ''

print("--------------------MENU---------------------")
for key in record.keys():
    print(key,":", record[key]['Name'],"\t|", record[key]['Price'],"\t|", record[key]['Qn'])
print("---------------------------------------------")
print('')

ui_name = input("Enter Your Name      : ")
ui_mail = input("Enter Mail ID        : ")
ui_ph = input("Enter Phone No      : ")
ui_pr = input("Enter product ID       : ")
ui_qn = int(input("Enter Quantiry     : "))

if ui_qn<=record[ui_pr]['Qn'] & record[ui_pr]['Qn']!=0 :
    print("---------------------------------------------")
    print("                   Bill                      ")
    print('')
    print("Name      : ", record[ui_pr]["Name"])
    print("Price (Rs): ", record[ui_pr]["Price"])
    print("Quantity  : ", ui_qn)
    bill = ui_qn*int(record[ui_pr]['Price'])
    discount = 0
    if bill>=5000:
        discount = bill/10
        print("---------------------------------------------")
        print("Billing   : ", bill-discount, "Rs")
        print("---------------------------------------------")
    else:
        print("---------------------------------------------")
        print("Billing   : ", bill-discount, "Rs")
        print("---------------------------------------------")

    record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn 
    print('')
    print("---------------------------------------------")
    print("  Thanks for your order, Inventory Updated!  ")
    print("---------------------------------------------")
    user_in = ui_name+","+ui_mail+","+ui_ph+","+record[ui_pr]['Name']+","+str(ui_qn*record[ui_pr]['Price'])+","+str(discount)+","+time.ctime()+"\n"
    
            
elif record[ui_pr]['Qn']==0:
    print('-'*50+"\n")
    print(f"Sorry we are having the Quantity of {record[ui_pr]['Name']} : {record[ui_pr]['Qn']} ")
    print('-'*50)
else:
    print("Sorry We are not Having Enough Quantity")
    print(f"We're only having {record[ui_pr]['Qn']} Quantity")
    print("If You Want To Purchase Press Y..")
    ch = input("Press Y/N")
    if ch.lower()=='y':
        print("---------------------------------------------")
        print("                   Bill                      ")
        print('')
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["Price"])
        print("Quantity  : ", record[ui_pr]['Qn'])
        bill = int(record[ui_pr]['Qn'])*int(record[ui_pr]['Price'])
        discount = 0
        if bill>=5000:
            discount = bill/10
            print("---------------------------------------------")
            print("Billing   : ", bill-discount, "Rs")
            print("---------------------------------------------")
        else:
            print("---------------------------------------------")
            print("Billing   : ", bill-discount, "Rs")
            print("---------------------------------------------")
        
         
        print('')
        print("---------------------------------------------")
        print("  Thanks for your order, Inventory Updated!  ")
        print("---------------------------------------------")
        user_in = ui_name+","+ui_mail+","+ui_ph+","+record[ui_pr]['Name']+","+str(int(record[ui_pr]['Qn'])*int(record[ui_pr]['Price']))+","+str(discount)+time.ctime()+"\n"
        record[ui_pr]['Qn'] = 0
        
    else:
        print("Thanks!!")
        

js = json.dumps(record)

fd = open("Records.json",'w')
fd.write(js)
fd.close()

fd = open('Sales_json.csv','a')
fd.write(user_in)
fd.close()
```
